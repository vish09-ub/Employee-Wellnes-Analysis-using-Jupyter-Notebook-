```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Treatment by Gender
sns.countplot(x='Gender', hue='treatment', data=df)
plt.title("Treatment by Gender")
plt.show()

# Family History vs Treatment
sns.countplot(x='family_history', hue='treatment', data=df)
plt.title("Family History vs Treatment")
plt.show()
```

```python
# Crosstab: Seek Help vs Treatment
pd.crosstab(df['seek_help'], df['treatment'], normalize='index') * 100

# Grouped stats
df.groupby('mental_vs_physical')['treatment'].value_counts(normalize=True)
```
